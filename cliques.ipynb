```python
def clique_graph(g, cmax=1000):
    ite = nx.find_cliques(g)
    i = 0
    cliques = []
    K = nx.Graph()
    while True:
        try:
            cli = next(ite)
            cliques.append(tuple(cli))
            i = i+1
            if i > cmax:
                return None
        except StopIteration:
            break
    # Add a numbered node for each clique.
    K.add_nodes_from(cliques)
    # Join cliques by an edge if they share a node.
    clique_pairs = itert.combinations(cliques, 2)
    K.add_edges_from((c1, c2) for (c1,c2) in clique_pairs if set(c1).intersection(set(c2)))
    return K
```

```python
kg = clique_graph(g)
kg.order()
```

```python
k2g=clique_graph(kg)
k2g.order()
```

```python
k3g=clique_graph(k2g)
k3g.order()
```

```python
k4g=clique_graph(k3g,500)
k4g.order()
```
