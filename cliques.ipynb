```python
import pycliques
import networkx as nx
from pycliques.cliques import clique_graph
from pycliques.small import list7

from networkx.drawing.nx_agraph import graphviz_layout, to_agraph

options = {
    'node_color': 'cyan',
    'node_size': 250,
    'line_color': 'grey',
    'linewidths': 0,
    'width': 1,
    'with_labels': True,
}
```

```python
g=clique_graph(nx.octahedral_graph())
```

```python
nx.draw(g,**options)
```

```python
A = to_agraph(g)
A.layout('dot')
A.draw('abcd.png')
```
